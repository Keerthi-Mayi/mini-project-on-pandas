```python
import pandas as pd
data=pd.read_csv("Sales_Data_Challenges.csv")
data.head()
```

```python
# 1. Calculate the total revenue generated by each city
total_revenue_by_city = data.groupby('City')['Revenue'].sum()
total_revenue_by_city
```

```python
# 2. Identify the category with the highest sales and most significant contribution to total revenue
category_summary = data.groupby('Category')[['Revenue', 'Quantity']].sum().sort_values(by='Revenue')
category_summary
```

```python
# 3. Find products with frequent sales discrepancies (e.g., high discount percentages, missing prices)
sales_discrepancies= data[(data['Price'].isna()) | (data['Discount'] > 70)]
print(sales_discrepancies.shape)
sales_discrepancies
```

```python
# 4. Identify the most preferred payment method
preferred_payment_method = data['PaymentMethod'].value_counts()
preferred_payment_method
```

```python
# 5.Check for cities with bias toward specific payment methods
city_payment_bias = data.groupby(['City', 'PaymentMethod']).size().unstack(fill_value=0)
city_payment_bias
```
